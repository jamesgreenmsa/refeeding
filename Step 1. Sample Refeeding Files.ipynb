emar Sample File - 10,000 rows

```python
import pandas as pd

# Path to the large CSV file
csv_file_path = '/content/drive/MyDrive/Research/Refeeding/hosp/emar.csv'

# Initialize an empty DataFrame to hold the sample
sample_data = []

# Load the CSV file in chunks
chunk_size = 10000  # Number of rows per chunk
for chunk in pd.read_csv(csv_file_path, chunksize=chunk_size):
    # Append the first 10,000 rows from the first chunk
    sample_data.append(chunk)

    # Stop after collecting enough samples
    if len(sample_data) * chunk_size >= 10000:
        break

# Concatenate the list of DataFrames into a single DataFrame
df_sample = pd.concat(sample_data)[:10000]  # Ensure it contains only 10,000 rows

# Save to a new CSV file
df_sample.to_csv('/content/drive/MyDrive/Research/Refeeding/hosp/emar_samp.csv', index=False)

# Display data types
print(df_sample.dtypes)
```
